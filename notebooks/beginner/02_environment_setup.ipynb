```python
# Let's start by checking your current Python environment
import sys
import platform
import subprocess

print("üêç Python Environment Check")
print("=" * 40)
print(f"Python Version: {sys.version}")
print(f"Platform: {platform.platform()}")
print(f"Architecture: {platform.architecture()[0]}")
print(f"Python Executable: {sys.executable}")
print()

# Check if we're in a virtual environment
in_venv = hasattr(sys, 'real_prefix') or (hasattr(sys, 'base_prefix') and sys.base_prefix != sys.prefix)
print(f"Virtual Environment: {'‚úÖ Active' if in_venv else '‚ùå Not active'}")

if not in_venv:
    print("‚ö†Ô∏è  RECOMMENDATION: Use a virtual environment for MCP development!")
    print("   This keeps your MCP dependencies separate from your system Python.")

print()
print("üì¶ Checking for pip...")
try:
    import pip
    print(f"‚úÖ pip version: {pip.__version__}")
except ImportError:
    print("‚ùå pip not found - please install pip first!")

# Check Python version compatibility
major, minor = sys.version_info[:2]
if major >= 3 and minor >= 8:
    print(f"‚úÖ Python {major}.{minor} is compatible with MCP")
else:
    print(f"‚ùå Python {major}.{minor} is too old - MCP requires Python 3.8+")
    print("   Please upgrade your Python installation.")
```

```python
# Let's install the core MCP dependencies
# NOTE: In a real environment, you'd run these commands in your terminal

install_commands = [
    "pip install --upgrade pip",
    "pip install mcp>=1.0.0",
    "pip install uvicorn>=0.24.0", 
    "pip install fastapi>=0.104.0",
    "pip install pydantic>=2.5.0",
    "pip install httpx>=0.25.0"
]

print("üì¶ Core MCP Dependencies Installation")
print("=" * 50)
print()
print("Run these commands in your terminal (with virtual environment active):")
print()

for i, cmd in enumerate(install_commands, 1):
    print(f"{i}. {cmd}")

print()
print("üí° Pro tip: You can also install all at once:")
print("   pip install mcp uvicorn fastapi pydantic httpx")
print()

# Let's try to import MCP to see if it's available
print("üîç Checking if MCP is already installed...")
try:
    import mcp
    print(f"‚úÖ MCP is installed! Version: {mcp.__version__ if hasattr(mcp, '__version__') else 'Unknown'}")
except ImportError:
    print("‚ùå MCP not found - you'll need to install it following the commands above")
    
try:
    import fastapi
    print(f"‚úÖ FastAPI is available! Version: {fastapi.__version__}")
except ImportError:
    print("‚ùå FastAPI not found - needed for HTTP-based MCP servers")

try:
    import uvicorn
    print("‚úÖ Uvicorn is available! (ASGI server for running MCP servers)")
except ImportError:
    print("‚ùå Uvicorn not found - needed for running MCP servers")
```
